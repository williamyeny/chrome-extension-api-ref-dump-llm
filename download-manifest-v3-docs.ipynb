```python
import requests
from bs4 import BeautifulSoup
import html2text

# Step 1: Fetch list of relevant links
def get_links():
    url = "https://developer.chrome.com/docs/extensions/reference/api"
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, "html.parser")
    links = []
    for a in soup.find_all('a', href=True):
        href = a['href']
        if href.startswith("/docs/extensions/reference/api/"):
            links.append("https://developer.chrome.com" + href)
    return links

# Main
links = get_links()

raw_markdowns = []
h = html2text.HTML2Text()
h.ignore_links = True

for link in links:
    raw_markdowns.append(h.handle(requests.get(link).text))
```

```python
def truncate(s):
    if 'Except as otherwise noted,' in s:
        s = s.split('Except as otherwise noted,')[0]
    if '#  chrome.' in s:
        s = s.split('#  chrome.', 1)[1]
        s = '#  chrome.' + s
    return s

markdowns = []
for raw_markdown in raw_markdowns:
    markdowns.append(truncate(raw_markdown))

print(markdowns[0])
```

```python
import json

data = [{link: markdown} for link, markdown in zip(links, markdowns)]

with open('output.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
